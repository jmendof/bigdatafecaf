#Modelo


- Iniciarlizar os pesos e bias
-Feedforward
-Calcular a perda
-Backpropagtion

```python
#Inicializar

w= np.random.randn(1)
b=0
```

```python
def forward(inputs, w, b):
  return inputs*w + b
```

```python
def mse(Y, y):
  return (Y-y)**2
```

```python
def backpropagation(inputs, outputs, targets, w, b, learning_rate):
  dw = learning_rate*(-2*inputs*(targets-outputs)).mean()
  db = learning_rate*(-2*(targets-outputs)).mean()

  w-= dw
  b-= db
  return w, b
```

```python
def model_fit(inputs, targets, w, b, learning_rate = 0.001, epochs = 200):
  for epoch in range(epochs):
    outputs = forward(inputs, w, b)
    loss = np.mean(mse(outputs, targets))
    w, b = backpropagation(inputs, outputs, targets, w, b, learning_rate)

    if (epoch+1)%50 == 0:
      print(f'Epoch: {epoch+1}, Loss: {loss:.4f}')
  return w, b
```

```python
x = np.arange(-10,10,2)
Y = get_linear_curve(x, w = 1.8, b = 32, noise_scale=2)
```

```python
#Inicialização

w = np.random.randn(1)
b = np.zeros(1)
```

```python
w, b = model_fit(x, Y, w, b, learning_rate = 0.005, epochs = 2000)
print(f'w: {w[0]:.3f}, b: {b[0]:.3f}')
```
